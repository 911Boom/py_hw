```python
import mindspore as ms
import sqlite3
import numpy as np
```

```python
conn = sqlite3.connect('ssq.db')
cur = conn.cursor()
```

```python
data_sql = 'select id,red1,red2,red3,red4,red5,red6,blue from ssq'
cur.execute(data_sql)
data = cur.fetchall()
data
```

```python
import mindspore as ms
import numpy as np
net = ms.nn.LSTM(10, 16, 2, has_bias=True, batch_first=True, bidirectional=False)
x = ms.Tensor(np.ones([3, 5, 10]).astype(np.float32))
h0 = ms.Tensor(np.ones([1 * 2, 3, 16]).astype(np.float32))
c0 = ms.Tensor(np.ones([1 * 2, 3, 16]).astype(np.float32))
output, (hn, cn) = net(x, (h0, c0))
print(x)
print(h0)
print(c0)
print(output)
```
